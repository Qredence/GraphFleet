# GraphFleet Quickstart Guide

This notebook demonstrates how to get started with GraphFleet, a powerful document search and analysis tool built on top of GraphRAG.

## Prerequisites

Before running this notebook, make sure you have:
1. Installed GraphFleet: `pip install graphfleet`
2. Set up your environment variables in `.env`
3. Access to the example documents

```python
# Import required libraries
from pathlib import Path
from graphfleet.core import GraphFleet

# Initialize project
project_dir = Path("./data")
GraphFleet.init_project(project_dir)
```

## Adding Documents

Let's add some example documents to our knowledge base. GraphFleet supports various document formats including TXT, PDF, and DOCX.

```python
# Create GraphFleet instance
gf = GraphFleet(project_dir)

# Add example documents
docs_dir = Path("./sample_docs")
for doc_path in docs_dir.glob("*.txt"):
    gf.add_document(doc_path)

print(f"Added {len(list(docs_dir.glob('*.txt')))} documents to the knowledge base")
```

## Basic Search

Now that we have some documents in our knowledge base, let's try searching through them using natural language queries.

```python
# Perform a simple search
query = "What are the main challenges in machine learning?"
results = gf.search(query)

# Display results
for i, result in enumerate(results, 1):
    print(f"\nResult {i} (Score: {result.score:.2f})")
    print(f"Document: {result.document.title}")
    print(f"Content: {result.content}")
```

## Advanced Features

GraphFleet supports advanced search features like filters and custom ranking.

```python
# Search with filters and options
results = gf.search(
    query="deep learning applications in healthcare",
    filters={"domain": "healthcare"},
    options={
        "max_results": 5,
        "min_score": 0.5
    }
)

# Display filtered results
for result in results:
    print(f"\nScore: {result.score:.2f}")
    print(f"Document: {result.document.title}")
    print(f"Content: {result.content}")
```

## Finding Similar Documents

GraphFleet can also find documents similar to a given document.

```python
# Get the first document as an example
doc = next(gf.list_documents())

# Find similar documents
similar_docs = gf.find_similar(doc.id)

print(f"\nDocuments similar to '{doc.title}':")
for similar in similar_docs:
    print(f"\nScore: {similar.score:.2f}")
    print(f"Document: {similar.document.title}")
```
