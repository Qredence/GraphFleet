This notebook demonstrates how to ingest documents into GraphFleet for processing and analysis.

```python
# Import required packages
from pathlib import Path
import json

# Note: graphfleet package needs to be installed first
from graphfleet import GraphFleet
```

## Initialize GraphFleet

First, we'll create a new project directory and initialize GraphFleet.

```python
# Create a project directory
project_dir = Path("./my_graph_project")
project_dir.mkdir(exist_ok=True)

# Initialize GraphFleet
graph_fleet = GraphFleet(str(project_dir))
```

## Prepare Sample Documents

Let's create some sample documents to ingest.

```python
# Sample documents
documents = [
    {
        "content": "GraphFleet is a powerful graph-based knowledge management system.",
        "metadata": {"type": "documentation", "category": "overview"}
    },
    {
        "content": "The system uses advanced NLP techniques for document processing.",
        "metadata": {"type": "technical", "category": "architecture"}
    }
]

# Save documents to files
docs_dir = project_dir / "documents"
docs_dir.mkdir(exist_ok=True)

for i, doc in enumerate(documents):
    with open(docs_dir / f"doc_{i}.json", "w") as f:
        json.dump(doc, f)
```

## Configure Ingestion Settings

Now we'll configure how documents should be processed during ingestion.

```python
# Configure ingestion settings
settings = {
    "chunk_strategy": "sentence",  # Can be 'sentence', 'fixed', or 'semantic'
    "chunk_size": 1000,
    "chunk_overlap": 200,
    "embedding_model": "text-embedding-3-large"
}
```

## Ingest Documents

Now we can ingest the documents into GraphFleet.

```python
# Ingest documents
result = graph_fleet.ingest_documents(documents, settings)
print(f"Processed {result['processed_count']} documents")
if result['failed_documents']:
    print(f"Failed to process {len(result['failed_documents'])} documents")
```

## Verify Ingestion

Let's verify that our documents were properly ingested by performing a simple search.

```python
# Perform a test search
search_results = graph_fleet.semantic_search(
    query_text="What is GraphFleet?",
    k=2,
    threshold=0.5
)

for result in search_results['results']:
    print(f"Score: {result['score']}")
    print(f"Content: {result['content']}\n")
```
