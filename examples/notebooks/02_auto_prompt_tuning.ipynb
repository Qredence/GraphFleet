# Auto Prompt Tuning with GraphFleet

This notebook demonstrates how to use GraphFleet's auto prompt tuning capabilities to optimize prompts for your specific use case.

```python
from graphfleet import GraphFleet
from graphfleet.app.models import PromptTuningRequest
from pathlib import Path
```

## Initialize GraphFleet

First, we'll load our existing project with ingested documents.

```python
project_dir = Path("./my_graph_project")
graph_fleet = GraphFleet(str(project_dir))
```

## Define Base Prompts

Let's define some base prompts that we want to tune.

```python
# Define base prompts for different tasks
base_prompts = {
    "summarization": "Summarize the following text: {text}",
    "entity_extraction": "Extract key entities from the following text: {text}",
    "relationship_mapping": "Identify relationships between entities in the text: {text}"
}
```

## Configure Tuning Parameters

Now we'll set up the parameters for prompt tuning.

```python
# Configure tuning parameters
tuning_config = {
    "model": "gpt-4",
    "optimization_metric": "relevance",  # Can be 'relevance', 'conciseness', or 'accuracy'
    "max_iterations": 5,
    "evaluation_samples": 3  # Number of samples to use for evaluation
}

# Create tuning request
tuning_request = PromptTuningRequest(
    base_prompts=base_prompts,
    config=tuning_config
)
```

## Run Prompt Tuning

Let's start the auto prompt tuning process.

```python
# Run prompt tuning
tuning_results = graph_fleet.tune_prompts(tuning_request)

# Display results
print("Tuned Prompts:")
for task, result in tuning_results.items():
    print(f"\n{task}:")
    print(f"Original: {base_prompts[task]}")
    print(f"Tuned: {result['prompt']}")
    print(f"Improvement: {result['improvement_score']:.2f}%")
```

## Test Tuned Prompts

Let's test our tuned prompts on some sample text.

```python
# Test sample
test_text = "GraphFleet utilizes advanced NLP techniques for document processing and knowledge extraction. The system maintains relationships between documents through a graph-based architecture."

# Test each tuned prompt
for task, result in tuning_results.items():
    tuned_prompt = result['prompt'].format(text=test_text)
    response = graph_fleet.generate_response(tuned_prompt)
    
    print(f"\n{task} Results:")
    print(response)
```
