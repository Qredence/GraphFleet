# Document Indexing with GraphFleet

This notebook demonstrates how to create and manage document indexes in GraphFleet.

```python
from graphfleet import GraphFleet
from graphfleet.core.types import IndexConfig, ChunkStrategy, EmbeddingConfig
from pathlib import Path
```

## Initialize GraphFleet

First, we'll load our existing project with ingested documents.

```python
project_dir = Path("./my_graph_project")
graph_fleet = GraphFleet(str(project_dir))
```

## Configure Indexing

Let's set up the configuration for our document index.

```python
# Create index configuration
index_config = IndexConfig(
    chunk_size=1000,
    chunk_overlap=200,
    chunk_strategy=ChunkStrategy.SENTENCE,
    embedding=EmbeddingConfig(
        model="text-embedding-3-large",
        dimension=3072,
        normalize=True,
        batch_size=32
    ),
    options={
        "similarity_metric": "cosine",
        "index_type": "hnsw",
        "store_document_text": True
    }
)
```

## Create Index

Now we'll create the index using our configuration.

```python
# Create the index
graph_fleet.create_index(**index_config.dict())
print("Index created successfully")
```

## Test Index with Semantic Search

Let's test our index by performing some semantic searches.

```python
# Perform semantic search
search_queries = [
    "What is GraphFleet's architecture?",
    "How does document processing work?",
    "Explain the graph-based system"
]

for query in search_queries:
    print(f"\nQuery: {query}")
    results = graph_fleet.semantic_search(
        query_text=query,
        k=3,
        threshold=0.5
    )
    
    for result in results['results']:
        print(f"Score: {result['score']:.3f}")
        print(f"Content: {result['content']}\n")
```

## Index Statistics

Let's examine some statistics about our index.

```python
# Get index statistics
stats = graph_fleet.get_index_stats()

print("Index Statistics:")
print(f"Total Documents: {stats['total_documents']}")
print(f"Total Chunks: {stats['total_chunks']}")
print(f"Average Chunk Size: {stats['avg_chunk_size']:.2f} characters")
print(f"Index Size: {stats['index_size_mb']:.2f} MB")
```
